```python
from pylab import*
from matplotlib import*
```

```python
x=linspace(0,20,50) 
y=sin(x)
plot(x,y,'--',color='teal')
xlabel('x')
ylabel(r'$\lambda$')
pylab.legend(loc='upper right')
savefig('sinx.pdf')
show()
```

```python
x=linspace(-pi,pi,100)
y=sin(x)+cos(x)
plot(x,y,'--',color='teal')
xlabel('x')
ylabel(r'$\lambda$')
pylab.legend(loc='upper right')
show()
```

```python
x=linspace(-10,10,100)
y=exp(-pow(x,2)/sqrt(pi))   
plot(x,y,'--',color='teal',label='r')
xlabel('x')
ylabel(r'$\lambda$')
pylab.legend(loc='upper right')
show()
```

```python
def g(x,y):
    return x**2+exp(x)+y
plot(x,g(x,2))
plot(x,sin(x))
ylim(0,5)
xlim(0,10)
show()
```
