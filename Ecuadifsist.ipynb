```python
from pylab import*
from matplotlib import*
from scipy import*
```

```python
def f(t,x,y):
    return 5*x-3*y
def g(t,x,y):
    return -6*x+2*y

N=100
a,b=0.,(1/4)
h=(b-a)/N
t,x,y=zeros(N),zeros(N),zeros(N)
t[0],x[0],y[0]=a,0.,1.

for i in range(1,N):
    k11=h*f(t[i-1],x[i-1],y[i-1])
    k12=h*g(t[i-1],x[i-1],y[i-1])
    
    k21=h*f(t[i-1]+0.5*h,x[i-1]+0.5*k11,y[i-1]+0.5*k12)
    k22=h*g(t[i-1]+0.5*h,x[i-1]+0.5*k11,y[i-1]+0.5*k12)
    
    k31=h*f(t[i-1]+0.5*h,x[i-1]+0.5*k21,y[i-1]+0.5*k22)
    k32=h*g(t[i-1]+0.5*h,x[i-1]+0.5*k21,y[i-1]+0.5*k22)
    
    k41=h*f(t[i-1]+h,x[i-1]+k31,y[i-1]+k32)
    k42=h*g(t[i-1]+h,x[i-1]+k31,y[i-1]+k32)
    
    x[i]=x[i-1]+(1/6)*(k11+2*k21+2*k31+k41)
    y[i]=y[i-1]+(1/6)*(k12+2*k22+2*k32+k42)
    t[i]=t[i-1]+h

    
    
plot(t,x,label='rkx')
plot(t,(1/3)*exp(-t)-(1/3)*exp(8*t),label='func1')
plot(t,y,label='rky')
plot(t,(2/3)*exp(-t)+(1/3)*exp(8*t),label='func2')
pylab.legend()
show()
```
