```python
#import plotly.plotly as py
import plotly.graph_objs as go
from chart_studio.plotly import iplot as py
# MatPlotlib
import matplotlib.pyplot as plt
from matplotlib import pylab

# Scientific libraries
from numpy import arange,array,ones
from scipy import stats


from pylab import*
from matplotlib import*
from scipy import*
```

```python
data = np.loadtxt('canicadata.txt.txt').T
xk=data[0]
fxk=data[1]
fk=data[2]

fk=(4/3)*pi*(fk/2)**3
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(xk,fxk)
line = slope*xk+intercept
slope2, intercept2, r_value2, p_value2, std_err2 = stats.linregress(xk,fk)
line2 = slope2*xk+intercept2


plt.figure(figsize=(10,10))
plt.plot(xk,fxk,'o',label='Datos experimentales mililitros')
plt.plot(xk,fk,'o',label='Datos experimentales centimetros cubicos')
plt.plot(xk,line,label='Mililitros')
plt.plot(xk,line2,label='Centimetros cubicos')
pylab.title('Regresion Lineal')
ax = plt.gca()
#ax.set_axis_bgcolor((0.898, 0.898, 0.898))
fig = plt.gcf()
plt.xlabel("g",fontsize=18)
plt.ylabel("ml/cm^3",fontsize=18)
plt.legend(loc='upper left', shadow=True, fontsize='x-large')
plt.savefig('canicasfeas.png')
#py.plot_mpl(fig, filename='Regresion Lineal')
```
